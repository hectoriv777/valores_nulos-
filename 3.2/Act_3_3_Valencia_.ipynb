```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importar archivo csv_Valencia
df_Valencia = pd.read_csv("Valencia_España_listings_50cols.csv")
df_Valencia
```

```python
#identificar valores nulos
null_counts = df_Valencia.isnull().sum()

# Filtrar para mostrar solo las columnas con valores nulos y su cantidad
cols_with_null_counts = null_counts[null_counts > 0]
print("Columnas con valores nulos y su cantidad:")
print(cols_with_null_counts)
```

**De 53 columnas que seleccionamos

```python
df_Valencia['neighbourhood'] = df_Valencia['neighbourhood'].fillna("Desconocido")
```

Rellenamos con "Desconocido" ya que, al ser el barrio/vecindario, cada dirección es única y no tendría sentido usar métodos como bfill o ffill. De esta forma mantenemos la fila y diferenciamos los casos sin información.

```python
df_Valencia['bathrooms'] = df_Valencia['bathrooms'].fillna(df_Valencia['bathrooms'].median())
bathrooms_median = df_Valencia['bathrooms'].median()
print("Mediana de 'bathrooms':", bathrooms_median)
```

En este caso es un valor numérico. Si un host no especificó el número de baños, lo más lógico es usar la mediana de la columna, ya que es más robusta frente a valores extremos (hay alojamientos con muchos baños que podrían distorsionar la media y generar outliers).

```python
df_Valencia['bathrooms_text'] = df_Valencia['bathrooms_text'].fillna("No especificado")
```

Como es texto, no tiene sentido calcular media o mediana. Lo mejor es rellenar con "No especificado", para marcar que el host no dio detalle.

```python
#identificar valores nulos
null_counts = df_Valencia.isnull().sum()

# Filtrar para mostrar solo las columnas con valores nulos y su cantidad
cols_with_null_counts = null_counts[null_counts > 0]
print("Columnas con valores nulos y su cantidad:")
print(cols_with_null_counts)
```

```python
df_Valencia["bedrooms"]= df_Valencia["bedrooms"].fillna(df_Valencia["bedrooms"].median())
```

En este caso es un valor numérico. Si un host no especificó el número de cuartos , lo más lógico es usar la mediana de la columna, ya que es más robusta frente a valores extremos (hay alojamientos con muchos cuartos que podrían distorsionar la media y generar outliers)

```python
df_Valencia["beds"]=df_Valencia["beds"].fillna(df_Valencia["beds"].median())
```

```python
#identificar valores nulos
null_counts = df_Valencia.isnull().sum()

# Filtrar para mostrar solo las columnas con valores nulos y su cantidad
cols_with_null_counts = null_counts[null_counts > 0]
print("Columnas con valores nulos y su cantidad:")
print(cols_with_null_counts)
```

```python
df_Valencia["price"] = df_Valencia["price"].fillna(df_Valencia["price"].median())
```

El precio es numérico y puede variar mucho según el alojamiento, por lo que la mediana es la mejor opción (más robusta ante precios extremos muy altos o muy bajos).

```python
df_Valencia["has_availability"] = df_Valencia["has_availability"].fillna("f")
```

Esta columna es booleana (Yes/No). Si está vacía, lo más razonable es asumir "f" (False / No disponible), ya que no podemos garantizar disponibilidad.

```python
df_Valencia["first_review"] = df_Valencia["first_review"].fillna("Sin reseña")
```

Esta columna es de tipo fecha. Cuando está vacía, significa que el anuncio nunca ha recibido una reseña. En ese caso lo más coherente es rellenar con "Sin reseña" para diferenciarlo, en lugar de inventar una fecha.

```python
df_Valencia["last_review"] = df_Valencia["last_review"].fillna("Sin reseña")
```

Al igual que con first_review, si no hay reseñas no podemos inventar una fecha. Lo más coherente es marcarlo como "Sin reseña".
